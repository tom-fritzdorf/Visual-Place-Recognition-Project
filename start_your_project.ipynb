```python
from pathlib import Path
import shutil

DRIVE_ZIP_DIR = Path("/content/drive/MyDrive/Visual-Place-Recognition-Project/data_zip")
LOCAL_DATA = Path("/content/data")
LOCAL_DATA.mkdir(parents=True, exist_ok=True)

for zip_path in DRIVE_ZIP_DIR.glob("*.zip"):
    dataset_name = zip_path.stem
    extract_dir = LOCAL_DATA / dataset_name

    if extract_dir.exists() and any(extract_dir.iterdir()):
        print(f"{dataset_name} already extracted locally, skipping.")
        continue

    print(f"Extracting {zip_path.name} to {extract_dir}...")
    extract_dir.mkdir(parents=True, exist_ok=True)
    shutil.unpack_archive(str(zip_path), str(extract_dir))
    print(f"Done extracting {dataset_name}")

print("All datasets extracted to:", LOCAL_DATA)
```
